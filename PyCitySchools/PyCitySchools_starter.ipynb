```python
# Dependencies and Setup
import pandas as pd

# Files to Load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate District Summary information

# Total schools
total_schools = school_data_complete['school_name'].nunique()
total_schools

# Total students
total_students = school_data_complete['Student ID'].nunique()
total_students

# Total budget
total_budget = school_data['budget'].sum()
total_budget

# Average math score
average_math = school_data_complete['math_score'].mean()
average_math

# Average reading score
average_read = school_data_complete['reading_score'].mean()
average_read

# Overall score
overall_score = (average_math + average_read) / 2
overall_score

# Percent passing math
passing_math = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
passing_math_count = passing_math['Student ID'].nunique()
passing_math_percent = ((passing_math_count / total_students) * 100)
passing_math_percent

# Percent passing reading
passing_read = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]
passing_read_count = passing_read['Student ID'].nunique()
passing_read_percent = ((passing_read_count / total_students) * 100)
passing_read_percent

# District summary table
district_summary_df = pd.DataFrame({'Total Schools': [total_schools], 'Total Students': [total_students], 
                                    'Total Budget': [total_budget], 'Average Math Score': [average_math],
                                    'Average Reading Score': [average_read], '% Passing Math': [passing_math_percent], 
                                    '% Passing Reading': [passing_read_percent], 
                                    '% Overall Passing Rate': [overall_score]})
district_summary_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
# Calculate School Summary information

# Group data by school, type and budget
schools_grouped = school_data_complete.groupby(['school_name', 'type', 'budget'])

# Average math scores by school
school_average_math = schools_grouped['math_score'].mean()
reset_index = school_average_math.reset_index()
school_average_math = reset_index['math_score']
school_average_math

# Average reading scores by school
school_average_read = schools_grouped[['reading_score']].mean()
reset_index = school_average_read.reset_index()
school_average_read = reset_index['reading_score']
school_average_read

# Percent pass by school
math_pass = school_data_complete.loc[school_data_complete['math_score'] >= 70, :]
school_math_pass = math_pass.groupby(['school_name'])
school_math_pass_percent = (school_math_pass['math_score'].count() / schools_grouped['math_score'].count() * 100)
reset_index = school_math_pass_percent.reset_index()
school_math_pass_percent = reset_index['math_score']
school_math_pass_percent

read_pass = school_data_complete.loc[school_data_complete['reading_score'] >= 70, :]
school_read_pass = read_pass.groupby(['school_name'])
school_read_pass_percent = (school_read_pass['reading_score'].count() / schools_grouped['reading_score'].count() * 100)
reset_index = school_read_pass_percent.reset_index()
school_read_pass_percent = reset_index['reading_score']
school_read_pass_percent

# Overall score
school_score = (((school_average_math) + (school_average_read)) / 2)
school_score

# Total students
school_students = schools_grouped['Student ID'].count()
reset_index = school_students.reset_index()
school_students = reset_index['Student ID']
school_students

# Reset index
schools_grouped = schools_grouped.count()
reset_index = schools_grouped.reset_index()

# School budget
school_budget = reset_index['budget']
school_budget

# Budget per student
per_student_budget = school_budget / school_students
per_student_budget

# School type
school_type = reset_index['type']
school_type

# School name
school_name = reset_index['school_name']
school_name

# School summary table
school_summary_df = pd.DataFrame({'School Name': school_name, 'School Type': school_type, 
                                  'Total Students': school_students, 'Total School Budget': school_budget, 
                                  'Per Student Budget': per_student_budget, 'Average Math Score': school_average_math, 
                                  'Average Reading Score': school_average_read, '% Passing Math': school_math_pass_percent, 
                                  '% Passing Reading': school_read_pass_percent, 'Overall Passing Rate': school_score
                                 })
school_summary_df
```
